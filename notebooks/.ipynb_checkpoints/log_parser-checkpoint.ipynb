```python
from tensorboard.backend.event_processing.event_accumulator import EventAccumulator
import os
import regex as re
import pandas as pd
import tqdm
import numpy as np
```

```python
LOGS_DIR = '../logs/'
REG_EDA = '\W*logs/(\d+)_(\d+)_BERT_(\w+)_alpha_0.(\d+)_num_aug_(\d+)_lr_(\d+)e-(\d+)_epochs_(\d+)_batch_size_(\d+)'
REG_TRAIN = '\W*logs/(\d+)_(\d+)_BERT_train_lr_(\d+)e-(\d+)_epochs_(\d+)_batch_size_(\d+)'
```

```python
event_acc = EventAccumulator('../logs/03_01_BERT_rd_alpha_0.1_num_aug_1_lr_2e-05_epochs_5_batch_size_16/')
event_acc.Reload()
# Show all tags in the log file
print(event_acc.Tags())
```

```python
log_dirs = [x[0] for x in os.walk(LOGS_DIR)][1:]
len(log_dirs)
```

```python
m = re.match('\W*logs/(\d+)_(\d+)_BERT_(\w+)_alpha_0.(\d+)_num_aug_(\d+)_lr_(\d+)e-(\d+)_epochs_(\d+)_batch_size_(\d+)', log_dirs[1])
date       = m.group(1) + '_' + m.group(2)
EDA_type   = m.group(3)
alpha      = int(m.group(4))/10
num_augs   = int(m.group(5))
lr         = int(m.group(6))*10**(-int(m.group(7)))
epochs     = int(m.group(8))
batch_size = int(m.group(9))
```

```python
column_names = ["date", "EDA_type", "alpha", "num_augs", "lr", "epochs", "batch_size", "accuracy"]
df = pd.DataFrame(columns = column_names)
df.head()
```

```python
df.loc[df.shape[0]] = [date, EDA_type, alpha, num_augs, lr, epochs, batch_size, 0]
df.head()
```

```python
# s is file name, df is current DF
def add_EDA_log(s, df):
    
    # get EDA info
    m = re.match(REG_EDA, s)
    if m:
        date       = m.group(1) + '_' + m.group(2)
        EDA_type   = m.group(3)
        alpha      = int(m.group(4))/10
        num_augs   = int(m.group(5))
        lr         = int(m.group(6))*10**(-int(m.group(7)))
        epochs     = int(m.group(8))
        batch_size = int(m.group(9))
        
        # get loss
        event_acc = EventAccumulator(s)
        event_acc.Reload()
        loss = event_acc.Scalars('Epoch_Validation_Accuracy')[-1][-1]
        
        # append
        df.loc[df.shape[0]] = [date, EDA_type, alpha, 
                               num_augs, lr, epochs, 
                               batch_size, loss]
```

```python
# s is file name, df is current DF
def add_TRAIN_log(s, df):
    
    # get EDA info
    m = re.match(REG_TRAIN, s)
    if m:
        date       = m.group(1) + '_' + m.group(2)
        EDA_type   = "Train"
        alpha      = np.nan
        num_augs   = np.nan
        lr         = int(m.group(3))*10**(-int(m.group(4)))
        epochs     = int(m.group(5))
        batch_size = int(m.group(6))
        
        # get loss
        event_acc = EventAccumulator(s)
        event_acc.Reload()
        loss = event_acc.Scalars('Epoch_Validation_Accuracy')[-1][-1]
        
        # append
        df.loc[df.shape[0]] = [date, EDA_type, alpha, 
                               num_augs, lr, epochs, 
                               batch_size, loss]
```

```python
add_TRAIN_log(log_dirs[12], df)
df.head()
```

```python
for s in log_dirs:
    add_EDA_log(s, df)
# for s in log_dirs:
#     add_TRAIN_log(s, df)
```
